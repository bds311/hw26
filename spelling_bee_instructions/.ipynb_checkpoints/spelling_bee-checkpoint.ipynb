# NYT Spelling Bee Re-creation

This notebook will outline instructions to complete a challenge problem which involves making a python re-creation of the new york times spelling bee game.  

Here is the link to the actual game for reference: https://www.nytimes.com/puzzles/spelling-bee

#### **Game Description:**  

The NYT spelling bee starts by giving the player 7 unique letters. One of these letters is highlighted as the "center letter".       
The goal is to find as many words as possible using the given letters.
- Words must be at least 4 letters long
- Words must contain the "center" letter
- Letters can be used more than once

The player earns points with each word guessed. Extra points are awarded if the guessded word is a `pangram`.  
In the context of the game, pangrams are words which use all 7 unique letters at least once.
  
    
#### **Selection process for 7 unique letters:**
To ensure that a pangram is possible for each selection of 7 unique letters, the 7 unique letters for the game are obtained by getting the set of letters from a randomly selected pangram.  
#### **Game Data:**
We provide english.txt which contains most english words (*not filtered for profanity*). Use this as the word bank for the game.

#### **Python Implementation**

We recommend implementing the game by printing text and asking for user inputs, although you do have the freedom to try an implement a graphical interface or any other method you see fit.

#### **Game Mechanics**

After selecting 7 unique letters (obtained from a randomly selected pangram within english.txt), these letters must be displayed to the player in some way. Additionally, one letter must be randomly selected to be the "center letter" and highlighted a different color than the other letters.

The player then must be able to input a guess and get feedback. Feedback includes...
- If the word is a valid word (uses center letter, is at leat 4 letters long, and only uses the given 7 letters), record the word in a list and display this list to the player to see which words they have already guessed.
    - If the word is a pangram using the 7 given letters, then let the player know they guessed a pangram.

- Give feedback on a wrong guess if the word is either 
    - Too short (less than 4 letters)
    - Doesn't use the center letter
    - Has letters not in the list of 7 given letters
    - not a real word (or at least not a word in english.txt)

#### **Meta**
Allow the player to quit or start a new game by inputing specific letters
